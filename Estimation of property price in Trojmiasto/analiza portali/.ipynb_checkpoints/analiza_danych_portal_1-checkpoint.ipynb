```python
def find_city(address):
    miasta = ['Gdańsk', 'Gdynia', 'Sopot']
    for m in miasta:
        if m in address:
            return m
    return 'Inne'

df['miasto']=df['adres'].apply(lambda x: find_city(x))
```
